```python
import os

folder = "/content/drive/MyDrive/Corpus-LST20/LST20_Corpus/train"

sentences = []
current = []

for fname in os.listdir(folder):
    path = os.path.join(folder, fname)

    with open(path, encoding="utf-8", errors="ignore") as f:
        for line in f:
            line = line.strip()

            if line == "":
                if current:
                    sentences.append(current)
                    current = []
                continue

            word = line.split()[0]
            current.append(word)

if current:
    sentences.append(current)
```

```python
texts = []
labels = []

for words in sentences:
    chars = []
    tags = []

    for word in words:
        if len(word) == 1:
            chars.append(word)
            tags.append("B")
        else:
            for i, c in enumerate(word):
                chars.append(c)
                if i == 0:
                    tags.append("B")
                elif i == len(word)-1:
                    tags.append("E")
                else:
                    tags.append("I")

    texts.append(chars)
    labels.append(tags)
```

```python
# char vocab
all_chars = set(c for sent in texts for c in sent)
char2id = {c:i+1 for i,c in enumerate(all_chars)}
char2id["<PAD>"] = 0

label2id = {"B":1, "I":2, "E":3, "<PAD>":0}
id2label = {v:k for k,v in label2id.items()}
```

```python
import torch
from torch.utils.data import Dataset

class MyDataset(Dataset):
    def __init__(self, texts, labels):
        self.texts = texts
        self.labels = labels

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        x = [char2id.get(c,0) for c in self.texts[idx]]
        y = [label2id[t] for t in self.labels[idx]]
        return torch.tensor(x), torch.tensor(y)
```

```python
from torch.nn.utils.rnn import pad_sequence

def collate_fn(batch):
    X, Y = zip(*batch)
    X = pad_sequence(X, batch_first=True)
    Y = pad_sequence(Y, batch_first=True)
    return X, Y
```

```python
import torch.nn as nn

class BiLSTM(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, 128, padding_idx=0)
        self.lstm = nn.LSTM(128, 256, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(512, 4)  # 4 classes

    def forward(self, x):
        x = self.embedding(x)
        x, _ = self.lstm(x)
        return self.fc(x)
```

```python
from torch.utils.data import DataLoader

dataset = MyDataset(texts, labels)
loader = DataLoader(dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)

model = BiLSTM(len(char2id))

loss_fn = nn.CrossEntropyLoss(ignore_index=0)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(8):
    model.train()
    total_loss = 0

    for X_batch, y_batch in loader:
        optimizer.zero_grad()

        outputs = model(X_batch)

        loss = loss_fn(
            outputs.view(-1, 4),
            y_batch.view(-1)
        )

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"epoch {epoch+1} | loss: {total_loss:.4f}")
```

```python
with open("/content/ws_test.txt", encoding="utf-8") as f:
    text = f.read()

import re
text = re.sub(r"\n\d+\s*", "\n", text)
text = text.replace("\n", "").replace(" ", "")

chars = list(text)

x = torch.tensor([[char2id.get(c,0) for c in chars]])

model.eval()
with torch.no_grad():
    outputs = model(x)
    pred_ids = outputs.argmax(dim=-1)[0]
```

```python
id2label_submit = {
    1: "B_WORD",
    2: "I_WORD",
    3: "E_WORD"
}

predictions = [
    id2label_submit[int(t)]
    for t in pred_ids
    if int(t) != 0
]
```

```python
import pandas as pd

sample = pd.read_csv("/content/ws_sample_submission.csv")

sample["Predicted"] = predictions

sample.to_csv("/content/submission.csv", index=False)
```
